# InstaShop

```python
df = pd.read_excel("Insta.xlsx")
df.info()
```

```python
df.sample(5)
```

```python
# switch value of amount to zero for payment type cash and card reader
df.loc[df["Payment Type"]== "Cash" , 'Amount'] = 0 
df.loc[df["Payment Type"]== "Card reader" , 'Amount'] = 0 
df.loc[df["Payment Type"]== "Card reader" , 'Service fee'] = 0 
df.loc[df["Payment Type"]== "Cash" , 'Service fee'] = 0 
```

```python

# column Instapoints fill nan values with zero
df["Instapoints"].fillna(0, inplace=True)
df["Service fee"].fillna(0, inplace=True)
```

```python
# Creat New column of Net Amount
df['NetAmount'] = df['Amount']+ df['Service fee'] + df ['Instapoints']
```

```python
# remove rows with zero value
df = df[df['NetAmount'] != 0]
```

```python
# split Submitted Date column to 3 columns
df['Submitted Date'] = pd.to_datetime(df['Submitted Date'], format='%m/%d/%Y %I:%M:%S %p')
df['Day'] = df['Submitted Date'].dt.day
df['Month'] = df['Submitted Date'].dt.month
df['Year'] = df['Submitted Date'].dt.year

```

```python
df.rename(columns={'Amount': 'InstaAmount'}, inplace=True)
```

```python
# Creat missing columns 
df['VOUCHER'] = 0
df['CHAIN'] = 0
df.rename(columns={'Site number': 'PROFIT'}, inplace=True)
df.rename(columns={'Site': 'BRANCH'}, inplace=True)
df.rename(columns={'Oder Code': 'ID'}, inplace=True)
df.rename(columns={'NetAmount': 'AMOUNT'}, inplace=True)
df.rename(columns={'Oder Code': 'ID'}, inplace=True)
df['CUSTOMER'] = 2016000040
df['CUSTOMER NAME'] = 0
df['BRANCH SAP'] = 0
df['CONTROLLER'] = 0
df['#'] = 'SITE'
df['SOURCE'] = 'SITE'
```

```python
df.head()
```

```python


# Assuming you have a DataFrame named 'df'
Insta = df[['ID', 'VOUCHER', 'Day', 'Month', 'Year', 'AMOUNT', 'CHAIN', 'BRANCH', 'CUSTOMER', 'PROFIT', 'CUSTOMER NAME', 'BRANCH SAP', 'CONTROLLER', '#', 'SOURCE']]
```
